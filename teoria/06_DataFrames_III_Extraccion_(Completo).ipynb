## Métodos `set_index` y `reset_index` 🔄
- El índice actúa como la colección de identificadores primarios/etiquetas/puntos de entrada para las filas.
- Extraer una fila desde un índice ordenado por posición o etiqueta es el método más rápido.
- Pandas utiliza las etiquetas/valores de índice al fusionar diferentes objetos.
- El método `set_index` establece una columna existente como índice del **DataFrame**.
- El método `reset_index` restablece el índice estándar numérico ascendente como índice del **DataFrame**.


```python
bond = pd.read_csv("jamesbond.csv", sep = ";")
bond.head()
```

```python
bond = bond.set_index("Film")
bond.head()
```

```python
bond = bond.reset_index()
bond
```

```python
bond = bond.reset_index().set_index("Year")
```

```python
bond = bond.set_index("Year")
```

## Recuperar Filas por Posición de Índice con `iloc` 🎯
- El accesor `iloc` recupera una o más filas por posición de índice.
- Usa un par de corchetes después del accesor.
- `iloc` acepta valores individuales, listas y slices.

```python
bond = pd.read_csv("jamesbond.csv")
bond.head()
```

```python
bond.iloc[5]
bond.iloc[[15, 20]]
bond.iloc[4:8]
bond.iloc[0:6]
bond.iloc[:6]
bond.iloc[20:] #Slicing -> traer paerte de la informacion
```

## Recuperar Filas por Etiqueta de Índice con `loc` 🔖
- El accesor `loc` recupera una o más filas por etiqueta de índice.
- Usa un par de corchetes después del accesor.

```python
bond = pd.read_csv("jamesbond.csv", sep = ";", index_col="Film")
bond.head()
```

```python
bond.loc["Goldfinger"]
```

```python
bond.loc["Dr. No"] # Etiqueta
```

```python
bond.iloc[0] # Posicion!
```

```python
bond.loc["Goldfinger"]
bond.loc["GoldenEye"]
bond.loc["Casino Royale"]
# bond.loc["Sacred Bond"]

bond.loc[["Octopussy", "Moonraker"]]
bond.loc[["Moonraker", "Octopussy"]]
bond.loc[["Moonraker", "Octopussy", "Casino Royale"]]
# bond.loc[["Moonraker", "Octopussy", "Casino Royale", "Gold Bond"]]

bond.loc["Diamonds Are Forever":"Moonraker"]
bond.loc["Moonraker":"Diamonds Are Forever"]

bond.loc["GoldenEye":] #Del registro Golden Eye hasta el final
bond.loc[:"On Her Majesty's Secret Service"] #Todos los reg del principio hasta "On her ..."


#Errores con un indice que no es unico!

# bond.loc[:"Casino Royale"]
# bond.loc["Casino Royale":]
```

## Segundo Argumento en los Accesores `loc` e `iloc` ➕
- El segundo valor dentro de los corchetes apunta a las columnas.
- `iloc` requiere posiciones numéricas para las filas y columnas.
- `loc` requiere etiquetas para las filas y columnas.

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film",sep=";").sort_index()
bond.head()
```

```python
bond.loc["Diamonds Are Forever", "Director"] # [Valor del Indice (etiqueta), Columna deseada]
```

```python
#Pasamos directamente la columna como 2 argumento
bond.loc["Diamonds Are Forever", "Director"]
bond.loc[["Octopussy", "GoldenEye"], "Director"]
bond.loc[["Octopussy", "GoldenEye"], "Director":"Budget"]
bond.loc["GoldenEye":"Octopussy", "Director":"Budget"]
bond.loc["GoldenEye":"Octopussy", ["Actor", "Bond Actor Salary", "Year"]]
```

```python
bond.head(1)
```

```python
bond.iloc[0, 2]
```

```python
bond.iloc[0, 2]
bond.iloc[3, 5]
bond.iloc[[0, 2], 3]
bond.iloc[[0, 2], [3, 5]]
bond.iloc[:7, :3]
```

## Sobrescribir un Valor en un DataFrame ✏️
- Usa los accesorios `iloc` o `loc` en el **DataFrame** para apuntar a un valor y asigna uno nuevo con el signo igual.


```python
bond = pd.read_csv("jamesbond.csv", index_col="Film", sep = ";").sort_index()
bond.head()
```

```python
#Duplicar el salario de "a view to a kill"
bond.loc["A View to a Kill","Bond Actor Salary"] = bond.loc["A View to a Kill","Bond Actor Salary"] * 2
```

```python
#Conversion de Sean a Caballero
bond.loc["Diamonds Are Forever", "Actor"]
bond.loc["Diamonds Are Forever", "Actor"] = "Sir Sean Connery"
```

```python
bond
```

## Sobrescribir Múltiples Valores en un DataFrame ✏️✏️
- El método `replace` reemplaza todas las ocurrencias de un valor en una **Series**.
- Para sobrescribir múltiples valores en un **DataFrame**, usa los accesorios `loc` o `iloc`.
- Estos accesorios pueden aceptar **Series** booleanas para apuntar a los valores específicos a sobrescribir.


```python
bond = pd.read_csv("jamesbond.csv", index_col="Film").sort_index()
bond.head()
```

```python
bond["Actor"] = bond["Actor"].replace("Sean Connery", "Sir Sean Connery") # (Valor que vamos a reemplazar, Valor nuevo)
```

```python
bond["Actor"] == "Sean Connery"
```

```python
bond[bond["Actor"] == "Sean Connery"].loc[:, "Actor"] = "Sir Sean Connery"
```

```python
is_sean_connery = bond["Actor"] == "Sean Connery"
bond.loc[is_sean_connery, "Actor"] = "Sir Sean Connery"
```

```python
bond
```

## Renombrar Etiquetas de Índice o Columnas en un DataFrame 🏷️
- El método `rename` acepta un diccionario para sus parámetros `columns` o `index`.
  - Las claves del diccionario representan los nombres existentes.
  - Los valores representan los nuevos nombres.
- También es posible sobrescribir directamente el atributo `columns` del **DataFrame**.


```python
bond = pd.read_csv("jamesbond.csv", index_col="Film").sort_index()
bond.head()
```

```python
bond = bond.rename(columns={ "Year": "Year of Release", "Box Office": "Revenue" })
bond.head()
```

```python
#Cambio de valores en el indice
swaps = {
    "Dr. No": "Dr No",
    "GoldenEye": "Golden Eye",
    "The World Is Not Enough": "Best Bond Movie Ever"
}

bond = bond.rename(index=swaps)
bond
```

```python
bond.columns = ["Year", "Bond Guy", "Camera Dude", "Revenues", "Cost", "Salary"] #Cambimos los nombres de la columna con una lista
```

## Crear una Muestra Aleatoria con el Método `sample` 🎲
- El método `sample` devuelve una o más filas aleatorias del **DataFrame**.
- Personaliza el parámetro `axis` para extraer columnas aleatorias.

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film", sep = ";").sort_index()
bond.head()
```

## Métodos `nsmallest` y `nlargest` 🔝🔽
- El método `nlargest` devuelve un número especificado de filas con los valores más grandes de una columna.
- El método `nsmallest` devuelve filas con los valores más pequeños de una columna.
- Ambos métodos son más eficientes que ordenar todo el **DataFrame**.

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film", sep = ";").sort_index()
bond.head()
```

```python
bond.sort_values("Box Office", ascending=False)
```

```python
top_5_peliculas_que_mas_ganaron = bond.nlargest(n = 5, columns="Box Office")
```

```python
top_5_peliculas_que_mas_ganaron
```

```python
bond.nlargest(n=4, columns="Box Office")
bond["Box Office"].nlargest(4)
```

```python
bond.nsmallest(3, columns="Bond Actor Salary")
```

```python
bond["Bond Actor Salary"].nsmallest(3)
```

```python
bond.nsmallest(3, columns="Bond Actor Salary")
bond["Bond Actor Salary"].nsmallest(3)
```

## Filtrar con el Método `where` 🔍
- Similar a los corchetes o `loc`, el método `where` filtra un **DataFrame** usando una **Series** booleana.
- Las filas que no cumplan con los criterios se rellenan con valores `NaN`.
- Mantener los valores `NaN` puede ser útil para ciertas operaciones de combinación y visualización.


```python
bond = pd.read_csv("jamesbond.csv", index_col="Film").sort_index()
bond.head()
```

```python
actor_is_sean_connery = bond["Actor"] == "Sean Connery"
actor_is_sean_connery
```

```python
actor_is_sean_connery = bond["Actor"] == "Sean Connery"
bond[actor_is_sean_connery]
bond.loc[actor_is_sean_connery]
bond.where(actor_is_sean_connery) # Si no cumple la condicion devuelve NaN / Nulo
```

## El Método `apply` con DataFrames 🛠️
- El método `apply` invoca una función en cada columna o fila del **DataFrame**.
- Pasa la función sin invocar como primer argumento.
- Usa el parámetro `axis` con el valor `"columns"` para aplicar la función a cada fila.
- Pandas pasará los valores de la fila como una **Series**, permitiendo el uso de accesorios como `loc` e `iloc`.

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film", sep = ";").sort_index()
bond.head()
```

```python
# Ejemplo de funcion

def rank_movie(row):
    year = row.loc["Year"]
    actor = row.loc["Actor"]
    budget = row.loc["Budget"]

    if year >= 1980 and year < 1990:
        return "Great 80's"

    if actor == "Pierce Brosnan":
        return "The best Bond ever!"

    if budget > 100:
        return "Expensive movie"

    return "No comment"

bond["rank_movie"] = bond.apply(rank_movie, axis = "columns")
```

```python

def budget_check(row):
    budget = row.loc["Budget"]

    if budget > 100:
        return "Expensive movie"

    if budget < 100:
        return "Normal"
    return "No comment"

bond["budget_flag"] = bond.apply(budget_check, axis = "columns")
```
