
### Conversión de Fechas y Tiempos 🕒
- El método `pd.to_datetime` convierte una **Series** en valores de tipo datetime.
- El parámetro `format` informa a pandas sobre el formato en que están almacenados los datos de tiempo.
- Usa símbolos para designar los segmentos de la cadena:
  - `%m`: Mes.
  - `%d`: Día.
  - '%p': PM/AM
- El atributo `dt` revela un objeto con múltiples atributos y métodos relacionados con fechas y tiempos.
  - El atributo `dt.time` extrae solo la hora de cada valor en una **Series** datetime.
- Usa el método `astype` para convertir los valores de una **Series** a otro tipo.
- El parámetro `parse_dates` de `read_csv` es otra forma de analizar cadenas como valores datetime.

## El Método `isin` ✅
- El método `isin` de una **Series** acepta un objeto de colección como una lista, tupla o **Series**.
- Devuelve `True` para una fila si su valor se encuentra en la colección.


```python
employees = pd.read_csv(path+"employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
legal_team = employees["Team"] == "Legal"
sales_team = employees["Team"] == "Sales"
product_team = employees["Team"] == "Product"

employees[legal_team | sales_team | product_team]
```

## El Método `duplicated` 🔁
- El método `duplicated` devuelve `True` si un valor de la **Series** es un duplicado.
- Marca una ocurrencia de un valor repetido como "no duplicado".
- Usa el parámetro `keep` para designar si la primera o última ocurrencia debe considerarse "no duplicada".
- Pasa `False` al parámetro `keep` para marcar todas las ocurrencias de valores repetidos como duplicados.
- Usa el símbolo tilde (`~`) para invertir los valores de una **Series**: `True` se convierte en `False` y viceversa.


```python
employees = pd.read_csv(path+"employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
employees["First Name"].duplicated()
```

```python
#Primero lee douglas, lo que va a devolver un falso, luego los proximos douglas devolvera un true
employees[employees["First Name"].duplicated()]
```

```python
employees[employees["First Name"].duplicated(keep="last")]
```

```python
employees[employees["First Name"].duplicated()]

#Cual consideramos? El primero o el ultimo?
employees[employees["First Name"].duplicated(keep="first")]

employees[employees["First Name"].duplicated(keep="last")]

employees[employees["First Name"].duplicated(keep=False)]

employees[~employees["First Name"].duplicated(keep=False)]
```

## El Método `drop_duplicates` 🗑️
- El método `drop_duplicates` elimina filas con valores duplicados.
- Por defecto, elimina una fila si *todos* sus valores se comparten con otra fila.
- El parámetro `subset` configura las columnas donde buscar valores duplicados.
- Pasa una lista al parámetro `subset` para buscar duplicados en múltiples columnas.


```python
employees = pd.read_csv(path +"employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
employees.drop_duplicates("Team")
```

```python
employees.drop_duplicates()

employees.drop_duplicates("Team")
employees.drop_duplicates("Team", keep="first")
employees.drop_duplicates("Team", keep="last")
employees.drop_duplicates("Team", keep=False)

employees.drop_duplicates("First Name", keep=False)

employees.drop_duplicates(["Senior Management", "Team"]).sort_values("Team")

employees.drop_duplicates(["Senior Management", "Team"], keep="last").sort_values("Team")
```
