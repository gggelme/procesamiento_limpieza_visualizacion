## Métodos Comunes de Cadenas en Series 🔡

## Parámetros `expand` y `n` del Método `split` ⚙️
- El parámetro `expand` devuelve un **DataFrame** en lugar de una **Series** de listas.
- El parámetro `n` limita el número de divisiones realizadas.

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head()
```

```python
chicago[["Last Name", "First Name"]] = chicago["Name"].str.split(",", expand=True)
```

```python
chicago.head()
```

```python
chicago[["Primary Title", "Secondary Title"]] = chicago["Position Title"].str.split(" ", expand=True, n=1)
```

## El Método `extract` 🎯
- El método `str.extract` utiliza expresiones regulares para extraer patrones de cadenas.
- Es útil para extraer información específica y estructurada de texto no estructurado.
- Puede extraer múltiples grupos de una expresión regular en columnas separadas de un **DataFrame**.

```python
data = {"columna_texto": ["ID: 123, Nombre: Juan", "ID: 456, Nombre: Maria", "ID: 789, Nombre: Pedro"]}
df_ejemplo = pd.DataFrame(data)

# Definir un patrón de expresión regular para extraer el ID y el Nombre
# r"..." indica una cadena sin procesar (raw string)
# (...) crea un grupo de captura para extraer una parte de la cadena
# ID:\s* Coincide con "ID:" seguido de cero o más espacios en blanco
# (\d+) Captura uno o más dígitos (el ID)
# ,\s* Coincide con una coma seguida de cero o más espacios en blanco
# Nombre:\s* Coincide con "Nombre:" seguido de cero o más espacios en blanco
# ([\w\s]+) Captura uno o más caracteres alfanuméricos o espacios en blanco (el Nombre)
pattern_simple = r"ID:\s*(\d+),\s*Nombre:\s*([\w\s]+)"

# Usar str.extract() para extraer el ID y el Nombre en nuevas columnas
# El método extract() devuelve un DataFrame con las partes capturadas
df_extraido = df_ejemplo["columna_texto"].str.extract(pattern_simple)

# Mostrar el DataFrame original y el DataFrame con los datos extraídos
print("DataFrame original:")
display(df_ejemplo)

print("\nDataFrame con datos extraídos:")
display(df_extraido)
```
