## ¿Qué es un DataFrame?
Un **DataFrame** es una estructura de datos bidimensional (tabla) que permite almacenar y manipular información en filas y columnas, similar a una hoja de cálculo de Excel.
- Cada columna puede tener un tipo de dato diferente (números, texto, fechas, etc.).
- Los valores faltantes se representan con `NaN` (*Not a Number*), y es importante saber cómo gestionarlos.
- Tanto las filas como las columnas tienen etiquetas (índices) que facilitan el acceso y manipulación de los datos.
- Pandas ofrece métodos comunes para Series y DataFrames, como `info()`, pero también métodos exclusivos para cada tipo.

## Diferencias entre métodos compartidos
Algunos métodos de Pandas funcionan tanto en Series como en DataFrames, pero su comportamiento puede variar:
- **`sum` en Series:** Suma todos los valores de la serie.
- **`sum` en DataFrame:** Por defecto suma los valores de cada columna (por filas).
Puedes usar el parámetro `axis` para cambiar el comportamiento:
- `axis=0` o "index" (comportamiento por defecto).
- `axis=1` o "columns"

Con el atributo `index_col` le podemos decir que columna usar de indice si viene. Si hay indices duplicados no falla, pero cambia la interaccion con el acceso por indices:
- Si el indice es unico, el acceso da una `Serie` de datos con indices las columnas del dataframe
- Si el indice no es unico, el resultado es un `Dataframe`

```python
revenue = pd.read_csv(datasets_path+"/revenue.csv", index_col="Date",sep=";")
#revenue_index = pd.read_csv(datasets_path+"/revenue.csv",sep = ";")
display(revenue)

print(type(revenue.loc["1/10/2026"]))
print(revenue.loc["1/10/2026"].index)
print(type(revenue.loc["1/1/2026"]))
print(revenue.loc["1/1/2026"].index)
```

Algunos metodos sobre series utilizando axis: Axis indica sobre las operacion que se va aplicar, lo que "desaparece" luego de aplicar la operacion
- Axis = "index" o axis=0, desaparecen los indices(renglones). La suma retorna una serie

```python
sum = revenue.sum(axis = 0)
display(sum)
print(type(sum))#serieeeeeeeeee

print(revenue.sum(axis = "columns", )) #o axis =1
```

## Selección de columnas en un DataFrame
- Para seleccionar una sola columna, puedes usar:
  - `df.columna` (solo si el nombre no tiene espacios).
  - `df["columna"]` (siempre funciona).
  Esto devuelve una **Series**.
- Para seleccionar varias columnas, usa una lista:
  - `df[["col1", "col2"]]`
  Esto devuelve un nuevo **DataFrame**.
- Si necesitas modificar los datos sin afectar el original, utiliza `.copy()` para crear una copia independiente.

_____________________

**Acceso a columnas**
Las columnas se pueden acceder con 
- Notacion punto `df.column`: Retorna serie con los datos
- Notacion corchetes `df[[column]]`: Podemos identificar mas de una columna. Si es una, es serie, si son varias es dataframe.


$$$$

**Acceso por indices:** Tenemos que acceder sabiendo los indices que tiene nuestro dataframe:
- df.loc[indice, columna]: usado como indice columna:
- df.iloc[posx, posy]: usado como pocision en el dataframe

## Manejo de valores faltantes: eliminar filas
Para limpiar tu DataFrame de valores faltantes, puedes usar el método `dropna`:
- Por defecto elimina filas con *cualquier* valor `NaN`.
- Usa el parámetro `how="all"` para eliminar filas donde *todos* los valores son `NaN`.
- Usa `subset` para especificar columnas a evaluar.
Esto ayuda a preparar los datos para análisis más precisos.

```python
nba = pd.read_csv(datasets_path + "/nba.csv", sep = ";")
nba.head()
```

```python
nba.info()

print(nba.shape)
```

```python
#nba.dropna(how = "any") elimina si falta algun dato
#nba.dropna(how="all") elimina si faltan todos los datos

nba.dropna(subset = ["Position", "Height"], how = "all") #elimino todos los que le faltan ambos

nba.dropna(subset = ["Salary"], how = "any")
```

## Rellenar valores faltantes con `fillna` (Imputacion de datos)
El método `fillna` reemplaza valores `NaN` con el argumento proporcionado.
- Puedes usarlo en **DataFrames** y **Series**.
- Devuelve una copia, no afecta el objeto original a menos que lo reasignes.
- Es útil para preparar los datos antes de análisis o visualización.


Observaciones: DEVUELVE COPIA, entonces es necesario indicarle el dataset para que aplique el cambio. Tiene inplace

```python
nba = pd.read_csv(datasets_path+"/nba.csv", sep = ";").dropna(how="all")
nba
#Con all no borramos las rows que contienen data + NaN
```

```python
#Reemplazamos los valores con 0 con fillna
nba_filled = nba.fillna(0) # no modifica el dataframe original, da una copia
display(nba_filled)

#Nos enfocamos en una sola columna y rellenamos con 0 los NaN
nba["Salary"] = nba["Salary"].fillna(nba["Salary"].mean())
nba["Team"] = nba["Team"].fillna("Por Asignar")

display(nba[nba.Team=="Por Asignar"])
#En el caso de college queremos poner "desconocido" y no un 0.0
nba["College"] = nba["College"].fillna(value="Unknown")
display(nba[nba.College=="Unknown"])

```

## Conversión de tipos de datos con `astype`
El método `astype` permite cambiar el tipo de datos de una columna, por ejemplo, de `float` a `int` o a `category`.
- Es importante asegurarse de que no haya valores faltantes antes de convertir.
- Usar el tipo `category` es útil para columnas con pocos valores únicos, optimizando el uso de memoria.
- El atributo `dtypes` muestra los tipos de las columnas en el DataFrame.


Observaciones: DEVUELVE COPIA, entonces es necesario indicarle el dataset para que aplique el cambio. No tiene inplace

```python
nba = pd.read_csv(datasets_path + "/nba.csv", sep = ";").dropna(how = "all") #Toda la fila debe tener N/A
nba["Salary"] = nba["Salary"].fillna(0)
nba["Weight"] = nba["Weight"].fillna(0)
nba
```

```python
display(nba.dtypes) #retorna una serie

print(nba.info())
```

```python
copia = nba["Salary"].astype("int") #genera siempre una copia!
copia.dtype #funciona
copia2=nba["Salary"].astype(int) #tambien da copia
copia2.dtype

nba["Salary"] = nba["Salary"].astype("int")
```

## El tipo `category` en Pandas
- El tipo `category` es ideal para columnas con pocos valores únicos, como posiciones o equipos.
- Usar `category` reduce el uso de memoria y mejora el rendimiento.
- Puedes contar los valores únicos con `nunique` y convertir una columna con `astype("category")`.


Observaciones: Los tipos de datos de categoria son mas pesados en memoria!!!

```python
nba = pd.read_csv(datasets_path + "/nba.csv", sep = ";").dropna(how = "all")
nba.tail()
```

```python
aa = nba["Team"].unique() #me da una lista de valores unicos
bb = nba.Team.nunique() #meda una lista de la cantidad de valroes unicos
print(bb)

nunique = nba.nunique() #Valores unicos por columna
display(nunique) # es una serie
```

```python
print(nba.info())
#Mirar la memory usage: 36.9+ KB

nba["Position"] = nba["Position"].astype("category")
nba["Team"] = nba["Team"].astype("category")

print()
print(nba.info())
```

## Ordenar un DataFrame con `sort_values`
Puedes ordenar tu DataFrame por una o varias columnas usando el método `sort_values`:
- Usa el parámetro `by` para especificar la columna o columnas.
- Personaliza el orden con `ascending` (True por defecto).
- Ajusta la posición de los `NaN` con `na_position`.

Observaciones. Le podemos pasar inplace para que no retorne una copia.

```python
nba = pd.read_csv(datasets_path + "/nba.csv", sep = ";")
```

```python
nba.sort_values(by = "Salary", ascending = False, na_position = "last", inplace = True)
# nba = nba.sort_values(by = "Salary", ascending = False, na_position = "last") Lo mismo que el inplace = True
display(nba)

nba = nba.sort_values(by=["Team","Salary"], ascending = True, na_position = "last") 
display(nba)

nba = nba.sort_values(by=["Team","Salary"], ascending = [True,False], na_position = "last")
display(nba)


# Como existe sort_values, tambien existe sort_index

nba = nba.sort_index(ascending=False)
nba
```

## Asignar rangos con el método `rank`
El método `rank` asigna un rango numérico a cada valor en una **Series**.
- Valores iguales comparten el mismo rango, dejando "huecos" en la numeración.
- Es útil para comparar posiciones relativas, por ejemplo, salarios dentro de un equipo.


Retorna una serie con los valroes con indices en la posicion de un ranking.  Si el rankin se repite, implica empate pero contabiliza posiiones.

```python
nba = pd.read_csv(datasets_path+"/nba.csv", sep =";")
nba = nba.sort_values(["Team", "Name"])
nba
```

```python
rank_serie = nba["Salary"].rank(ascending = True)
display(rank_serie)

#puedo ordenar acorde a este criterio

nba["ranking_salary"]=nba["Salary"].rank(ascending=False) #serie con indices originales y valroes posiciones. Uso las pocisiones como valores!
display(nba.sort_values(by = "ranking_salary", ascending = True))
```

## Funcion apply

Aplica una funcion como condicion a una columna

```python
#Ejemplo con condicionales de rango de salarios
def categorize_salary(salary):
    if salary > 20000000:
        return "Alto"
    elif salary > 5000000:
        return "Medio"
    else:
        return "Bajo"

nba["Salary Category"] = nba["Salary"].apply(categorize_salary)
nba.head()
```
